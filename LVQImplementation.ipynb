```python
import pandas as pd
from pandas import DataFrame
from math import sqrt
from random import randrange
```

```python
#membaca dataset 
def readData(filename):
    dataset = pd.read_csv(filename)
    print('Heart Disease Dataset')
    print(dataset)
    df = DataFrame(dataset)
    #membuat list dari dataset
    listDataset = df.values.tolist()
    return listDataset
```

```python
#menghitung  jarak antar data 
def euclidianDistance(data1, data2):
    euclidDist = 0.0
    for i in range (len(data1) - 1):
        euclidDist += (data1[i] - data2[i]) ** 2
    return sqrt(euclidDist)
```

```python
#memilih data secara random dari dataset sebagai codebook yang akan ditraining
def randomCodebook(dataset):
    n_data = len(dataset)
    n_fitur = len(dataset[0])
    codebook = [dataset[randrange(n_data)][i] for i in range(n_fitur)]
    return codebook
```

```python
#memilih jarak terdekat antara setiap codebook dan data
def bestMatch(codebookVector, data):
    listDist = list()
    for codebook in codebookVector:
        dist = euclidianDistance(codebook, data)
        listDist.append((codebook, dist))
    #mengurutkan data berdasarkan tuple jarak
    listDist.sort(key=lambda tup : tup[1])
    #mengembalikan data codebook dengan jarak terdekat
    return listDist[0][0]
```

```python
#melakukan training codebook terpilih
def trainCodebook(dataset, n_codebook, learnRate, total_epochs):
    codebookVector = [randomCodebook(dataset) for i in range(n_codebook)]
    print('Codebook Vector')
    print(codebookVector)
    print('')

    for epoch in range(total_epochs):
        print('------------------------------------------------------------------------')
        print('')
        print('Epoch ke - ', epoch)
        print('')

        #menghitung learning rate setiap epoch
        lRate = learnRate * (1.0 - (epoch / float(total_epochs)))
        print('Learn Rate Epoch ke - ', epoch, ' : ',lRate)
        print('')

        jumlahError = 0.0
        i_bmu = 1
        for data in dataset:
            bmu = bestMatch(codebookVector, data)
            print('BMU ke - ', i_bmu)
            print(bmu)
            print('')
            for i in range(len(data) - 1):
                error = data[i] - bmu[i]
                jumlahError += error ** 2
                #update bmu jika kelas sama atau berbeda dengan data
                if bmu[-1] == data[-1]:
                    bmu[i] += lRate * error
                else:
                    bmu[i] -= lRate * error 
            i_bmu += 1
        print('Jumlah Error pada Epoch ke - ', epoch, ' : ', jumlahError)
        print('')
```

```python
def run() :
    filename = 'HeartDisease.csv'
    ListDataset = readData(filename)

    print('')
    print('List Dataset')
    print(ListDataset)
    print('')

    learnRate = 0.3
    n_epochs = 10
    n_codebook = 2
    Codebooks = trainCodebook(ListDataset, n_codebook, learnRate, n_epochs)
   
```

```python
if __name__ == '__main__':
    run()
```
